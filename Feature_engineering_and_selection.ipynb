```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
pd.set_option('display.max_rows',None)
```

```python
train=pd.read_csv(r'/content/train.csv')
#printing top 5 rows
train.head(5)
```

## missing values

```python
missing_features=[i for i in train.columns if train[i].isnull().sum()>0]
for i in missing_features:
  print(i,np.round(train[i].isnull().mean(),4),"% values")
```

```python
for i in missing_features:
  if train[i].dtypes=='object':
    train[i]=train[i].fillna("missing_value")
  else:
    train[i]=train[i].fillna(train[i].median())
```

```python
train.isnull().sum()
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    train[feature]=train['YrSold']-train[feature]
```

```python
train.head()
```

```python
train[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

##numerical values
- numerical data is skewed so perform log normal distribution

```python

import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    train[feature]=np.log(train[feature])
```

```python
train.head(5)
```

## categorical values

```python
for i in train.columns:
  print(i,len(train[i].unique()))
```

```python
train.head()
```

##Handling Rare Categorical Feature

```python
categorical_features=train.select_dtypes(include=['object'])
```

```python
for feature in categorical_features:
    temp=train.groupby(feature)['SalePrice'].count()/len(train)
    temp_df=temp[temp>0.01].index
    train[feature]=np.where(train[feature].isin(temp_df),train[feature],'Rare_var')
```

```python
for feature in categorical_features:
    labels_ordered=train.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    train[feature]=train[feature].map(labels_ordered)
```

```python
train.head(5)
```

# Feature Engineering

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
X=train.drop(['Id','SalePrice'],axis=1)
scaler.fit(X)
X_scaled=scaler.transform(X)
```

```python
y=train['SalePrice']
```

```python
print(X.shape,y.shape)
```

- feature selection we will apply lasso model

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV,train_test_split,KFold
from sklearn.feature_selection import SelectFromModel
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(X_scaled, y)
```

```python

feature_sel_model.get_support()
```

```python
selected_feat = X.columns[(feature_sel_model.get_support())]
len(selected_feat)
```

```python
X=X[selected_feat]
```
